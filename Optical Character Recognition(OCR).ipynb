# Importing Libraries

```python
import easyocr
import cv2
from matplotlib import pyplot as plt
import numpy as np
from IPython.display import display, Image
```

# OCR on Single Word

```python
#Image 1 before OCR
image_path = 'stop.jpg'
im1 = cv2.imread('stop.jpg')
plt.imshow(im1)
plt.show()
```

```python
reader = easyocr.Reader(['en'], gpu=False)
result = reader.readtext(image_path)
result
```

```python
top_left = tuple(result[0][0][0])
bottom_right = tuple(result[0][0][2])
text = result[0][1]
fontFace = cv2.FONT_ITALIC
```

```python
#Image 1 after OCR
img = cv2.imread(image_path)
img = cv2.rectangle(img,top_left,bottom_right,(0,220,0),3)
img = cv2.putText(img,text,top_left, fontFace, 0.70,(220,0,0),2,cv2.FONT_ITALIC)
plt.figure(figsize=(6,6))
plt.imshow(img)
plt.show()
```

# OCR on Multiple Words

```python
#Image 2 before OCR
image_path2='good quote.jpg'
im2 = cv2.imread('good quote.jpg')
plt.figure(figsize=(6,6))
plt.imshow(im2)
plt.show()
```

```python
reader = easyocr.Reader(['en'], gpu=False)
result = reader.readtext(image_path2)
result
```

```python
#Image 2 after OCR
img = cv2.imread(image_path2)
for detection in result:
    top_left = tuple([int(val) for val in detection[0][0]])
    bottom_right = tuple([int(val) for val in detection[0][2]])
    text = detection[1]
    font = cv2.FONT_ITALIC
    img = cv2.rectangle(img,top_left,bottom_right,(0,255,0),2)
    img = cv2.putText(img,text,top_left, font,0.85,(0,0,210),2,cv2.LINE_AA)
    
plt.figure(figsize=(6,6))
plt.imshow(img)
plt.show()
```

# OCR on Multiple Lines

```python
#Image 3 before OCR
image_path3='multiple lines.jpg'
im3 = cv2.imread('multiple lines.jpg')
im3 = cv2.cvtColor(im3,cv2.COLOR_BGR2RGB)
plt.imshow(im3)
plt.show()
```

```python
reader = easyocr.Reader(['en'], gpu=False)
result = reader.readtext(image_path3)
result
```

```python
#Image 3 after OCR
img = cv2.imread(image_path3)
for detection in result:
    top_left = tuple([int(val) for val in detection[0][0]])
    bottom_right = tuple([int(val) for val in detection[0][2]])
    text = detection[1]
    font = cv2.FONT_HERSHEY_TRIPLEX
    img = cv2.rectangle(img,top_left,bottom_right,(0,255,0),1)
    img = cv2.putText(img,text,top_left, font,0.80,(0,0,255),2,cv2.LINE_AA)
    
plt.figure(figsize=(7,7))
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```
